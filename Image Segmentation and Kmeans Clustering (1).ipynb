# Image Segmentation and K means Clustering 

Image segmentation is the process of partitioning a digital image into multiple segments (sets of pixels, also known as image objects). The goal of segmentation is to simplify or change the representation of an image into something that is more meaningful and easier to analyze. Image segmentation is typically used to locate objects and boundaries (lines, curves, etc.) in images. More precisely, image segmentation is the process of assigning a label to every pixel in an image such that pixels with the same label share certain characteristics.

The result of image segmentation is a set of segments that collectively cover the entire image, or a set of contours extracted from the image. Each of the pixels in a region are similar with respect to some characteristic or computed property, such as color, intensity, or texture. Adjacent regions are significantly different color respect to the same characteristics. When applied to a stack of images, typical in medical imaging, the resulting contours after image segmentation can be used to create 3D reconstructions with the help of interpolation algorithms like marching cubes.

# Tiger.png

```python
# Importing 'pyplot' and 'imread' from matplotlib

from matplotlib.image import imread  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading png image file
tiger = imread('tiger.png')
```

```python
# show image
plt.figure(figsize = (14,10))
plt.imshow(tiger)
```

```python
### Looking at the Image (tiger) numpy array  ...
tiger
```

```python
### Getting the dimensions of the Image(tiger)...
tiger.shape
```

```python
### Reshaping the Image to a two dimensional array....
x = tiger.reshape(-1,3)
x.shape
```

```python
### Importing KMeans Cluster from sklearn.cluster and Creating an instance of Kmeans with K set = 16...
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 16)
```

```python
### Fitting the data on the KMeans cluster model
kmeans.fit(x)
```

```python
### Image segmentation => Kmaens cluster centers....
Img_sgmn = kmeans.cluster_centers_ 
Img_sgmn
```

```python
## Image labels ....
Image = Img_sgmn[kmeans.labels_]
Image
```

```python
### reshaping the array to original form of the imagr(3d)
Image = Image.reshape(tiger.shape)
```

```python
# show image
plt.figure(figsize = (14,10))
plt.imshow(Image)
```

# Dogs.jepg

```python
### Importing Image from PIL...
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
### Loading the Image ....
dogs = Image.open('dogs.jpeg')
```

```python
### show Image...

plt.figure(figsize = (14,10))
plt.imshow(dogs)
```

```python
### Converting the Image into numpy array ...
dogs_np = np.asarray(dogs)
dogs_np
```

```python
### Scaling ...
dogs_np = dogs_np/255.0
```

```python
### Getting the dimensions of the array ....
dogs_np.shape
```

```python
### Reshaping the dimensions to 2D array ...

dogs_np = dogs_np.reshape((-1,3))
dogs_np.shape
```

```python
### Creating an instance of Kmeans with K set = 3...

kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(dogs_np)
```

```python
### Predicting the pixel Labels ....
labels = kmeans.predict(dogs_np)
labels
```

```python
#### Reshaping the Labels (3D)....
labels = labels.reshape((185,272))
```

```python
#### Plotting the Image
plt.figure(figsize = (14,10))
plt.imshow(labels)
```

```python
### Getting the Cluster_centers 
centers = kmeans.cluster_centers_
centers
```

```python
### Plotting the cluster centers to get the three dominant colours of the Image coressponding to the three centers[0,1,2]...
blob = np.ones((100,100,3))
blob1 = blob * centers[0]
plt.imshow(blob1)
```

```python
### Plotting the colour corresponding to 2nd cluster...
blob2 = blob * centers[1]
plt.imshow(blob2)
```

```python
### Plotting the colour corresponding to 3rd cluster...
blob3 = blob * centers[2]
plt.imshow(blob3)
```
